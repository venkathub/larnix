```python
import pandas as pd

orders = pd.DataFrame({
    "item": ["pen", "bag", "lamp"],
    "price": ["₹1,200", "₹85", "₹2,499"],
})

orders["price_num"] = (
    orders["price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

print(orders)
print("total:", orders["price_num"].sum())
print("mean :", round(orders["price_num"].mean(), 2))
```
