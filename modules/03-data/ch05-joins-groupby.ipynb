```python
import pandas as pd

def group_totals(data, key, val):
    return pd.DataFrame(data).groupby(key)[val].sum().to_dict()

run_tests([("sum by group", group_totals({"g": ["a", "a", "b"], "v": [1, 2, 10]}, "g", "v"), {"a": 3, "b": 10})])
```

```python
import pandas as pd

def group_means(data, key, val):
    return pd.DataFrame(data).groupby(key)[val].mean().round(2).to_dict()

run_tests([("mean by group", group_means({"g": ["a", "a", "b"], "v": [2, 4, 10]}, "g", "v"), {"a": 3.0, "b": 10.0})])
```
