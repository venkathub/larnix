```python
def class_shares(counts):
    total = sum(counts.values())
    return {k: round(v / total, 2) for k, v in counts.items()}

run_tests([
    ("imbalanced", class_shares({"a": 3, "b": 1}), {"a": 0.75, "b": 0.25}),
    ("even", class_shares({"x": 1, "y": 1}), {"x": 0.5, "y": 0.5}),
])

run_tests([
    ("imbalanced", class_shares({"a": 3, "b": 1}), {"a": 0.75, "b": 0.25}),
    ("even", class_shares({"x": 1, "y": 1}), {"x": 0.5, "y": 0.5}),
])
```
