```python
def min_max_scale(values):
    lo = min(values)
    hi = max(values)
    return [(v - lo) / (hi - lo) for v in values]

run_tests([
    ("0..10", min_max_scale([0, 5, 10]), [0.0, 0.5, 1.0]),
    ("shifted", min_max_scale([2, 4, 6]), [0.0, 0.5, 1.0]),
])

run_tests([
    ("0..10", min_max_scale([0, 5, 10]), [0.0, 0.5, 1.0]),
    ("shifted", min_max_scale([2, 4, 6]), [0.0, 0.5, 1.0]),
])
```

```python
import numpy as np

def standardize(values):
    a = np.array(values, dtype=float)
    return [round(float((v - a.mean()) / a.std()), 2) for v in values]

run_tests([
    ("four values", standardize([2, 4, 6, 8]), [-1.34, -0.45, 0.45, 1.34]),
])

run_tests([
    ("four values", standardize([2, 4, 6, 8]), [-1.34, -0.45, 0.45, 1.34]),
])
```
