```python
import pandas as pd

def first_n(data, n):
    return int(pd.DataFrame(data).head(n).shape[0])

run_tests([
    ("first 2", first_n({"a": [1, 2, 3, 4]}, 2), 2),
    ("first 0", first_n({"a": [1, 2, 3, 4]}, 0), 0),
])
```

```python
import pandas as pd

def count_above(data, col, threshold):
    return int((pd.DataFrame(data)[col] > threshold).sum())

run_tests([
    ("two above 2", count_above({"x": [1, 2, 3, 4]}, "x", 2), 2),
    ("none above 9", count_above({"x": [1, 2, 3, 4]}, "x", 9), 0),
])
```
