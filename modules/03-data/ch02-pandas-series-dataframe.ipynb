```python
import pandas as pd

def series_total(values):
    return int(pd.Series(values).sum())

run_tests([
    ("sums", series_total([10, 20, 30]), 60),
    ("one value", series_total([7]), 7),
])
```

```python
import pandas as pd

def column_mean(data, col):
    return round(float(pd.DataFrame(data)[col].mean()), 2)

run_tests([
    ("mean of a", column_mean({"a": [1, 2, 3], "b": [4, 5, 6]}, "a"), 2.0),
    ("mean of b", column_mean({"a": [1, 2, 3], "b": [4, 5, 6]}, "b"), 5.0),
])
```
