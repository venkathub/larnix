```python
import pandas as pd

def count_missing(data):
    return int(pd.DataFrame(data).isna().sum().sum())

run_tests([
    ("two missing", count_missing({"a": [1, None, 3], "b": [None, 2, 3]}), 2),
    ("none missing", count_missing({"a": [1, 2], "b": [3, 4]}), 0),
])
```

```python
import pandas as pd

def drop_missing(data):
    return int(pd.DataFrame(data).dropna().shape[0])

run_tests([
    ("one row survives", drop_missing({"a": [1, None, 3], "b": [4, 5, None]}), 1),
    ("all survive", drop_missing({"a": [1, 2], "b": [3, 4]}), 2),
])
```
