```python
import pandas as pd

def long_row_count(data):
    return int(pd.DataFrame(data).melt().shape[0])

run_tests([
    ("2x2 wide", long_row_count({"a": [1, 2], "b": [3, 4]}), 4),
    ("one column", long_row_count({"x": [1, 2, 3]}), 3),
])
```

```python
import pandas as pd

def n_variables(data):
    return int(pd.DataFrame(data).melt()["variable"].nunique())

run_tests([
    ("three cols", n_variables({"a": [1], "b": [2], "c": [3]}), 3),
    ("one col", n_variables({"x": [1, 2]}), 1),
])
```
