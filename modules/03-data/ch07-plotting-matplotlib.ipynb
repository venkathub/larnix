```python
def squared(xs):
    return [x * x for x in xs]

run_tests([
    ("squares", squared([1, 2, 3]), [1, 4, 9]),
    ("with zero", squared([0, 5]), [0, 25]),
])
```

```python
from collections import Counter

def category_counts(items):
    return dict(Counter(items))

run_tests([
    ("counts", category_counts(["a", "a", "b"]), {"a": 2, "b": 1}),
    ("empty", category_counts([]), {}),
])
```
