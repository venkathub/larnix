```python
import numpy as np

def mean_and_std(values):
    a = np.array(values)
    return (round(float(a.mean()), 2), round(float(a.std()), 2))

run_tests([
    ("simple", mean_and_std([2, 4, 6, 8]), (5.0, 2.24)),
    ("flat", mean_and_std([5, 5, 5]), (5.0, 0.0)),
])
```

```python
def count_in_range(values, lo, hi):
    return int(sum(1 for v in values if lo <= v <= hi))

run_tests([
    ("middle bin", count_in_range([1, 2, 3, 4, 5], 2, 4), 3),
    ("none", count_in_range([1, 2, 3], 10, 20), 0),
])
```
