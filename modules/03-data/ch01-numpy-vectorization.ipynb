```python
import numpy as np

arr = np.array([1, 5, 3, 8, 2, 4])

mask = arr > 3
print("mask:     ", mask)        # True where the element exceeds 3
print("selected: ", arr[mask])   # keep only those elements
print("shortcut: ", arr[arr > 3])  # the same, in one step
```

```python
import numpy as np

arr = np.array([1, 5, 3, 8, 2, 4])

print("mask:           ", arr > 3)
print("how many > 3:   ", (arr > 3).sum())   # True counts as 1
print("their average:  ", arr[arr > 3].mean())
```

```python
import numpy as np

def scale(arr, factor):
    return (np.array(arr) * factor).tolist()

run_tests([
    ("double", scale([1, 2, 3], 2), [2, 4, 6]),
    ("by zero", scale([5, 9], 0), [0, 0]),
])
```

```python
import numpy as np

def count_above(arr, threshold):
    return int((np.array(arr) > threshold).sum())

run_tests([
    ("three above 3", count_above([1, 5, 3, 8, 2, 4], 3), 3),
    ("none above 5", count_above([1, 2, 3], 5), 0),
])
```
