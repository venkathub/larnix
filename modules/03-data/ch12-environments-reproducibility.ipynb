```python
import numpy as np

def is_reproducible(seed):
    a = np.random.default_rng(seed).integers(0, 100, size=5)
    b = np.random.default_rng(seed).integers(0, 100, size=5)
    return list(a) == list(b)

run_tests([
    ("seed 42", is_reproducible(42), True),
    ("seed 7", is_reproducible(7), True),
])
```
