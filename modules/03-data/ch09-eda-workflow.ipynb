```python
import pandas as pd

def eda_shape(data):
    return tuple(pd.DataFrame(data).shape)

run_tests([
    ("2x3", eda_shape({"a": [1, 2], "b": [3, 4], "c": [5, 6]}), (2, 3)),
    ("1 col", eda_shape({"a": [1, 2, 3]}), (3, 1)),
])
```

```python
import pandas as pd

def summary(data, col):
    df = pd.DataFrame(data)
    s = df[col]
    return {"mean": round(float(s.mean()), 2), "min": int(s.min()), "max": int(s.max())}

run_tests([
    ("col a", summary({"a": [1, 2, 3, 4]}, "a"), {"mean": 2.5, "min": 1, "max": 4}),
])
```
