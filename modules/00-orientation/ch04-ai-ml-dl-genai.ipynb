```python
from collections import defaultdict

corpus = [
    "the cat sat on the mat",
    "the dog sat on the rug",
    "the cat ran on the grass",
]

following = defaultdict(list)
for sentence in corpus:
    words = sentence.split()
    for current, nxt in zip(words, words[1:]):
        following[current].append(nxt)

print("After 'the', these words have appeared:", following["the"])
print("After 'sat', these words have appeared:", following["sat"])
```

```python
import random
random.seed(7)   # so your sentence matches this page

word = "the"
out = [word]
for _ in range(5):
    options = following.get(word)
    if not options:
        break
    word = random.choice(options)   # pick one word that has followed `word`
    out.append(word)

print("Generated:", " ".join(out))
```
