```python
timeline = [
    (1958, "Rosenblatt's Perceptron"),
    (2012, "AlexNet wins ImageNet"),
    (2017, "The transformer"),
]

def event_year(timeline, name):
    for year, event in timeline:
        if event == name:
            return year
    return None

run_tests([
    ("finds AlexNet's year", event_year(timeline, "AlexNet wins ImageNet"), 2012),
    ("finds the transformer's year", event_year(timeline, "The transformer"), 2017),
    ("unknown event is None", event_year(timeline, "Moon landing"), None),
])
```

```python
timeline = [
    (1958, "Perceptron"),
    (1986, "Backpropagation"),
    (2012, "AlexNet"),
    (2017, "Transformer"),
    (2022, "ChatGPT"),
]

def events_in_era(timeline, start, end):
    names = []
    for year, event in timeline:
        if start <= year <= end:
            names.append(event)
    return names

run_tests([
    ("the deep-learning decade", events_in_era(timeline, 2010, 2019), ["AlexNet", "Transformer"]),
    ("a single year", events_in_era(timeline, 2022, 2022), ["ChatGPT"]),
    ("an empty era", events_in_era(timeline, 1990, 2000), []),
])
```
