```python
messages = [
    "win a free prize now",
    "lunch at noon?",
    "free money click here",
    "call me later",
    "you won a free trip",
]
labels = [1, 0, 1, 0, 1]   # 1 = spam, 0 = not spam (ham)

for text, is_spam in zip(messages, labels):
    print(("SPAM" if is_spam else "ham "), "|", text)
```

```python
from collections import Counter

spam_words, ham_words = Counter(), Counter()
for text, is_spam in zip(messages, labels):
    for word in set(text.split()):
        (spam_words if is_spam else ham_words).update([word])

# Words that show up in spam but never in the normal messages.
learned = sorted(w for w in spam_words if w not in ham_words)
print("Words the program learned to associate with spam:")
print(learned)
```

```python
def looks_spammy(message, spam_words):
    return any(word in spam_words for word in message.split())

new_message = "claim your free prize"
print(new_message, "->", "spam" if looks_spammy(new_message, learned) else "ham")
```

```python
def count_with_word(messages, word):
    return sum(1 for m in messages if word in m)

run_tests([
    ("counts the matches", count_with_word(["win free money", "hello", "free lunch"], "free"), 2),
    ("none match", count_with_word(["a", "b"], "free"), 0),
])
```

```python
def classify(message, spam_words):
    for word in spam_words:
        if word in message:
            return "spam"
    return "ham"

run_tests([
    ("a spammy message", classify("win a free prize", ["free", "win"]), "spam"),
    ("a clean message",  classify("see you at noon", ["free", "win"]), "ham"),
    ("no spam words to check", classify("anything goes", []), "ham"),
])
```
