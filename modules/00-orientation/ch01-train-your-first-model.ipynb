```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
acc = accuracy_score(y_test, model.predict(X_test))
print(f'Worked-example accuracy: {acc:.2%}')
assert acc > 0.8, 'expected a reasonable accuracy on Iris'
```

```python
def _is_number(x):
    return isinstance(x, (int, float)) and not isinstance(x, bool)

def _equalish(got, expected, tol):
    if _is_number(got) and _is_number(expected):
        return abs(got - expected) <= tol
    return got == expected

def run_tests(tests, tol=1e-9):
    tests = list(tests)
    failures = []
    for label, got, expected in tests:
        ok = _equalish(got, expected, tol)
        if not ok:
            failures.append(label)
    if failures:
        raise AssertionError('failed: ' + ', '.join(failures))
    print(f'All {len(tests)} tests passed')
    return True
```

```python
# Exercise 1 solution: fit() trains the model.
model.fit(X_train, y_train)
assert len(model.predict(X_test)) == len(y_test)

# Exercise 2 solution: accuracy().
def accuracy(y_true, y_pred):
    correct = sum(1 for a, b in zip(y_true, y_pred) if a == b)
    return correct / len(y_true)

run_tests([
    ('all correct', accuracy([1, 1, 1], [1, 1, 1]), 1.0),
    ('half right',  accuracy([1, 0],    [1, 1]),    0.5),
    ('none right',  accuracy([0, 0],    [1, 1]),    0.0),
])
print('Twin checks passed.')
```
