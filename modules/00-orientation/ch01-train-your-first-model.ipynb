```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
print("Measurements for", X.shape[0], "flowers, with", X.shape[1], "numbers each.")
print("First flower's measurements:", X[0])
print("Its species (as a number 0/1/2):", y[0])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
print("Training flowers:", X_train.shape[0])
print("Test flowers:", X_test.shape[0])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
print("Trained! The model has now seen", X_train.shape[0], "labelled flowers.")
```

```python
predictions = model.predict(X_test)
print("Predicted species for the first 10 test flowers:")
print(predictions[:10])
print("The true species for those same flowers:")
print(y_test[:10])
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, model.predict(X_test))
print(f"Accuracy on the test set: {acc:.2%}")
```

```python
model.fit(X_train, y_train)

run_tests([
    ("model predicts one label per test flower", len(model.predict(X_test)), len(y_test)),
])
```

```python
def accuracy(y_true, y_pred):
    correct = sum(1 for a, b in zip(y_true, y_pred) if a == b)
    return correct / len(y_true)

run_tests([
    ("all correct", accuracy([1, 1, 1], [1, 1, 1]), 1.0),
    ("half right",  accuracy([1, 0],    [1, 1]),    0.5),
    ("none right",  accuracy([0, 0],    [1, 1]),    0.0),
])
```
