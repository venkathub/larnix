```python
def posterior(true_pos, false_pos):
    return true_pos / (true_pos + false_pos)

run_tests([
    ("disease example", round(posterior(99, 495), 4), 0.1667),
    ("balanced", round(posterior(50, 50), 4), 0.5),
])
```

```python
def bayes(prevalence, sensitivity, false_positive_rate):
    true_pos = prevalence * sensitivity
    false_pos = (1 - prevalence) * false_positive_rate
    return true_pos / (true_pos + false_pos)

run_tests([
    ("rare disease", round(bayes(0.01, 0.99, 0.05), 4), 0.1667),
    ("common", round(bayes(0.5, 0.99, 0.05), 4), 0.9519),
])
```
