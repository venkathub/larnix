```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)            # fixed seed → the same draws every run
true_mean = 50.0

# Draw 500 separate samples, each of 40 values, from the same population,
# and record the mean of each sample.
sample_means = [rng.normal(loc=true_mean, scale=10, size=40).mean()
                for _ in range(500)]

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(sample_means, bins=30, color="tab:blue", edgecolor="white")
ax.axvline(true_mean, color="tab:red", linewidth=2, label="true mean = 50")
ax.set_xlabel("sample mean")
ax.set_ylabel("count (samples per bin)")
ax.set_title("500 sample means → a narrow bell around the truth")
ax.legend()
plt.show()

print("average of the sample means:", round(float(np.mean(sample_means)), 2))
print("spread of the sample means :", round(float(np.std(sample_means)), 2))
print("formula standard error std/sqrt(n):", round(10 / np.sqrt(40), 2))
```

```python
def sample_mean(sample):
    return sum(sample) / len(sample)

run_tests([
    ("three values", sample_mean([2, 4, 6]), 4.0),
    ("four values", sample_mean([10, 20, 30, 40]), 25.0),
])
```

```python
import numpy as np

def standard_error(sample):
    return np.std(sample) / np.sqrt(len(sample))

run_tests([
    ("spread sample", round(standard_error([2, 4, 6, 8]), 4), 1.118),
    ("no spread", round(standard_error([5, 5, 5, 5]), 4), 0.0),
])
```
