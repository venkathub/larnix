```python
import matplotlib.pyplot as plt

# Corners of a unit square, traced as a closed loop for plotting.
square = [(0, 0), (1, 0), (1, 1), (0, 1), (0, 0)]

def apply(M, pts):
    (a, b), (c, d) = M
    return [(a * x + b * y, c * x + d * y) for (x, y) in pts]

A = [[2, 0], [0, 1]]    # stretch horizontally (x doubles)
B = [[1, 0], [0, -1]]   # flip vertically (y flips sign)

after_A = apply(A, square)
after_B = apply(B, after_A)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(*zip(*square), label="original")
ax.plot(*zip(*after_A), label="after A (stretch)")
ax.plot(*zip(*after_B), label="after B (then flip)")
ax.axhline(0, color="gray", linewidth=0.5)
ax.axvline(0, color="gray", linewidth=0.5)
ax.set_aspect("equal")
ax.legend()
ax.set_title("A, then B")
plt.show()
```

```python
def matmul(X, Y):
    # entry [i][j] = dot product of row i of X with column j of Y
    return [[X[i][0] * Y[0][j] + X[i][1] * Y[1][j] for j in range(2)]
            for i in range(2)]

A = [[2, 0], [0, 1]]    # stretch horizontally
B = [[1, 0], [0, -1]]   # flip vertically
combined = matmul(B, A)  # "A then B" as one matrix
print("B·A =", combined)

x, y = 1, 1
ax_, ay = A[0][0] * x + A[0][1] * y, A[1][0] * x + A[1][1] * y       # after A
bx, by = B[0][0] * ax_ + B[0][1] * ay, B[1][0] * ax_ + B[1][1] * ay  # then B
cx, cy = combined[0][0] * x + combined[0][1] * y, combined[1][0] * x + combined[1][1] * y
print("A then B:", (bx, by), "| combined at once:", (cx, cy))
```

```python
def matmul2(A, B):
    def dot(row, col):
        return row[0] * col[0] + row[1] * col[1]

    cols = [[B[0][0], B[1][0]], [B[0][1], B[1][1]]]
    return [
        [dot(A[0], cols[0]), dot(A[0], cols[1])],
        [dot(A[1], cols[0]), dot(A[1], cols[1])],
    ]

run_tests([
    ("identity leaves B", matmul2([[1, 0], [0, 1]], [[2, 3], [4, 5]]), [[2, 3], [4, 5]]),
    ("double then ones", matmul2([[2, 0], [0, 2]], [[1, 1], [1, 1]]), [[2, 2], [2, 2]]),
])

run_tests([
    ("identity leaves B", matmul2([[1, 0], [0, 1]], [[2, 3], [4, 5]]), [[2, 3], [4, 5]]),
    ("double then ones", matmul2([[2, 0], [0, 2]], [[1, 1], [1, 1]]), [[2, 2], [2, 2]]),
])
```
