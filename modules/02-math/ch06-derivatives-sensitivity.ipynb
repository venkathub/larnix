```python
import matplotlib.pyplot as plt

def f(x):
    return x * x

# The curve, sampled densely so it looks smooth.
xs = [i / 20 for i in range(-20, 61)]   # -1.0 … 3.0
ys = [f(x) for x in xs]

# The tangent at x = 1: slope 2, passing through (1, f(1)).
x0 = 1.0
slope_at_x0 = 2 * x0                    # the derivative of x^2 is 2x
tangent = [f(x0) + slope_at_x0 * (x - x0) for x in xs]

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(xs, ys, color="tab:blue", label="f(x) = x²")
ax.plot(xs, tangent, "--", color="tab:orange", label="tangent at x=1 (slope 2)")
ax.scatter([x0], [f(x0)], color="tab:red", zorder=5)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("A derivative is the local slope")
ax.legend()
plt.show()
```

```python
def slope(p1, p2):
    rise = p2[1] - p1[1]
    run = p2[0] - p1[0]
    return rise / run

run_tests([
    ("through (0,0) and (2,6)", slope((0, 0), (2, 6)), 3.0),
    ("through (1,1) and (3,5)", slope((1, 1), (3, 5)), 2.0),
])
```

```python
def numeric_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

run_tests([
    ("slope of 2x+1 is 2", numeric_derivative(lambda x: 2 * x + 1, 5), 2.0),
    ("slope of 3x is 3", numeric_derivative(lambda x: 3 * x, 0), 3.0),
    ("slope of a constant is 0", numeric_derivative(lambda x: 7, 4), 0.0),
])
```
