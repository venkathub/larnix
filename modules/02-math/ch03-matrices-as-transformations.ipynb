```python
import matplotlib.pyplot as plt

# Corners of the unit square; repeat the first corner to close the outline.
square = [[0, 0], [1, 0], [1, 1], [0, 1], [0, 0]]

# A stretch matrix: double the x-direction, leave y alone.
M = [[2, 0], [0, 1]]

def apply(m, v):
    x, y = v[0], v[1]
    return [m[0][0]*x + m[0][1]*y, m[1][0]*x + m[1][1]*y]

moved = [apply(M, p) for p in square]

before_x = [p[0] for p in square]
before_y = [p[1] for p in square]
after_x = [p[0] for p in moved]
after_y = [p[1] for p in moved]

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(before_x, before_y, "o-", color="tab:blue", label="before")
ax.plot(after_x, after_y, "s--", color="tab:orange", label="after [[2,0],[0,1]]")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("A matrix moves the square")
ax.set_aspect("equal")
ax.legend()
plt.show()
```

```python
def stretch_x(v, factor):
    return [v[0] * factor, v[1]]

run_tests([
    ("double width", stretch_x([3, 4], 2), [6, 4]),
    ("five times", stretch_x([1, 1], 5), [5, 1]),
    ("factor 1 keeps it", stretch_x([2, 2], 1), [2, 2]),
])
```
