```python
import matplotlib.pyplot as plt

values = [2, 4, 6, 8, 10]

mean = sum(values) / len(values)
variance = sum((v - mean) ** 2 for v in values) / len(values)
std = variance ** 0.5

fig, ax = plt.subplots(figsize=(5, 3))
# The values as dots on a number line (all at the same height).
ax.scatter(values, [0] * len(values), color="tab:orange", zorder=5, label="values")
# The mean as a vertical line.
ax.axvline(mean, color="tab:blue", linestyle="--", label=f"mean = {mean}")
# The spread: one standard deviation either side of the mean.
ax.axvspan(mean - std, mean + std, alpha=0.15, color="tab:blue",
           label=f"±1 std ≈ {round(std, 2)}")
ax.set_yticks([])
ax.set_xlabel("value")
ax.set_title("Mean and standard deviation")
ax.legend(loc="upper center", bbox_to_anchor=(0.5, 1.35), ncol=3, fontsize=8)
plt.show()
```

```python
def mean(values):
    return sum(values) / len(values)

run_tests([
    ("mean of [2, 4, 6]", mean([2, 4, 6]), 4.0),
    ("all the same", mean([5, 5, 5]), 5.0),
    ("two values", mean([0, 10]), 5.0),
])
```

```python
def variance(values):
    m = sum(values) / len(values)
    squared_deviations = [(v - m) ** 2 for v in values]
    return sum(squared_deviations) / len(values)

run_tests([
    ("variance of [2, 4, 6]", variance([2, 4, 6]), 8 / 3),
    ("all the same", variance([5, 5, 5]), 0.0),
    ("two values", variance([0, 10]), 25.0),
])
```
