```python
import matplotlib.pyplot as plt

def f(x):
    return x * x

def gradient(x):
    return 2 * x          # slope of x^2 at x

lr = 0.2                  # learning rate (step size)
x = 9.0                   # where the ball starts
xs = [x]
for _ in range(8):
    x = x - lr * gradient(x)   # one downhill step
    xs.append(x)

# The bowl itself, drawn smoothly across the range.
curve_x = [v / 10 for v in range(-100, 101)]
curve_y = [f(v) for v in curve_x]

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(curve_x, curve_y, color="gray", label="f(x) = x^2")
ax.plot(xs, [f(v) for v in xs], "o-", color="tab:blue", label="descent steps")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("Rolling downhill to the minimum")
ax.legend()
plt.show()

print("x after each step:", [round(v, 3) for v in xs])
```

```python
def descent_step(x, grad, lr):
    return x - lr * grad

run_tests([
    ("step from 10", descent_step(10, 4, 0.1), 9.6),
    ("negative move", descent_step(0, 5, 0.1), -0.5),
    ("zero gradient = no move", descent_step(3, 0, 0.1), 3),
])

run_tests([
    ("step from 10", descent_step(10, 4, 0.1), 9.6),
    ("negative move", descent_step(0, 5, 0.1), -0.5),
    ("zero gradient = no move", descent_step(3, 0, 0.1), 3),
])
```

```python
def step_x_squared(x, lr):
    grad = 2 * x
    return x - lr * grad

run_tests([
    ("from 5, lr 0.1", step_x_squared(5, 0.1), 4.0),
    ("from 4, lr 0.5", step_x_squared(4, 0.5), 0.0),
    ("already at minimum", step_x_squared(0, 0.1), 0.0),
])

run_tests([
    ("from 5, lr 0.1", step_x_squared(5, 0.1), 4.0),
    ("from 4, lr 0.5", step_x_squared(4, 0.5), 0.0),
    ("already at minimum", step_x_squared(0, 0.1), 0.0),
])
```
