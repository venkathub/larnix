```python
def dot(a, b):
    return sum(x * y for x, y in zip(a, b))

ana = [5, 1, 0]
bo = [4, 2, 0]
cleo = [0, 1, 5]

print("Ana · Bo   =", dot(ana, bo), "(big positive -> similar)")
print("Ana · Cleo =", dot(ana, cleo), "(near zero -> unrelated)")
```

```python
def dot(a, b):
    return a[0] * b[0] + a[1] * b[1] + a[2] * b[2]

run_tests([
    ("dot([1,2,3],[4,5,6])", dot([1, 2, 3], [4, 5, 6]), 32),
    ("dot([1,0,0],[0,1,0])", dot([1, 0, 0], [0, 1, 0]), 0),
])

run_tests([
    ("dot([1,2,3],[4,5,6])", dot([1, 2, 3], [4, 5, 6]), 32),
    ("dot([1,0,0],[0,1,0])", dot([1, 0, 0], [0, 1, 0]), 0),
])
```

```python
def aligned(a, b):
    dot = sum(x * y for x, y in zip(a, b))
    return dot > 0

run_tests([
    ("shared taste", aligned([1, 2], [3, 4]), True),
    ("opposite", aligned([1, 0], [-1, 0]), False),
    ("unrelated", aligned([1, 0], [0, 1]), False),
])

run_tests([
    ("shared taste", aligned([1, 2], [3, 4]), True),     # dot = 11
    ("opposite", aligned([1, 0], [-1, 0]), False),       # dot = -1
    ("unrelated", aligned([1, 0], [0, 1]), False),       # dot = 0, not > 0
])
```
