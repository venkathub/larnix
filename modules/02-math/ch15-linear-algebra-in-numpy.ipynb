```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a * b   =", a * b)        # [ 4 10 18 ]  -> element-wise, still a vector
print("sum     =", np.sum(a * b))# 32           -> add them up by hand
print("np.dot  =", np.dot(a, b)) # 32           -> the dot product directly
print("a @ b   =", a @ b)        # 32           -> the same, with the @ operator
```

```python
import numpy as np

# A matrix that scales x by 2 and y by 3.
M = np.array([[2, 0],
              [0, 3]])
v = np.array([4, 5])

result = M @ v
print("M @ v   =", result)       # [8 15] -> x doubled, y tripled
print("shape   =", result.shape) # (2,)   -> still a 2-D point
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("A @ B =")
print(A @ B)                     # [[19 22] [43 50]]
print("shape =", (A @ B).shape)  # (2, 2)
```

```python
import numpy as np

M = np.array([[1, 2, 3],
              [4, 5, 6]])

print("M.shape   =", M.shape)    # (2, 3)
print("M.T =")
print(M.T)                       # [[1 4] [2 5] [3 6]]
print("M.T.shape =", M.T.shape)  # (3, 2)
```

```python
import numpy as np

I = np.eye(2)
v = np.array([5, 7])

print("I =")
print(I)                         # [[1. 0.] [0. 1.]]
print("I @ v =", I @ v)          # [5. 7.] -> v, unchanged
```

```python
import numpy as np

def dot(a, b):
    return int(np.array(a) @ np.array(b))

run_tests([
    ("3-vectors", dot([1, 2, 3], [4, 5, 6]), 32),
    ("orthogonal", dot([1, 0], [0, 1]), 0),
])
```

```python
import numpy as np

def matvec(M, v):
    return (np.array(M) @ np.array(v)).tolist()

run_tests([
    ("identity keeps v", matvec([[1, 0], [0, 1]], [5, 7]), [5, 7]),
    ("scales axes", matvec([[2, 0], [0, 3]], [4, 5]), [8, 15]),
])
```
