```python
import numpy as np
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6, 7, 8]
score = [52, 55, 60, 64, 69, 72, 79, 84]

r = np.corrcoef(hours, score)[0, 1]
print("correlation r =", round(r, 4))

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(hours, score)
ax.set_xlabel("hours studied")
ax.set_ylabel("exam score")
ax.set_title(f"Study hours vs score (r = {round(r, 2)})")
plt.show()
```

```python
import numpy as np

def correlation(x, y):
    return np.corrcoef(x, y)[0, 1]

run_tests([
    ("perfect positive", round(correlation([1, 2, 3, 4], [2, 4, 6, 8]), 4), 1.0),
    ("perfect negative", round(correlation([1, 2, 3, 4], [8, 6, 4, 2]), 4), -1.0),
])
```

```python
import numpy as np

def covariance(x, y):
    x = np.array(x)
    y = np.array(y)
    return np.mean((x - np.mean(x)) * (y - np.mean(y)))

run_tests([
    ("rises together", round(covariance([1, 2, 3, 4], [1, 2, 3, 4]), 4), 1.25),
    ("moves opposite", round(covariance([1, 2, 3, 4], [4, 3, 2, 1]), 4), -1.25),
])
```
