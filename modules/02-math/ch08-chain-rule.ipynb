```python
import matplotlib.pyplot as plt

def inner(x):
    return 2 * x          # u = 2x, inner slope 2

def outer(u):
    return u * u          # y = u^2, outer slope 2u

def composed(x):
    return outer(inner(x))   # y = (2x)^2

# The chain rule: outer slope (2u) times inner slope (2), at x0.
x0 = 1.0
u0 = inner(x0)
chain_slope = (2 * u0) * 2               # = 8 * x0

# A no-calculus cross-check: the central difference from Ch6.
h = 1e-5
numeric_slope = (composed(x0 + h) - composed(x0 - h)) / (2 * h)
print(f"chain-rule slope at x={x0}: {chain_slope}")
print(f"numeric slope at x={x0}:    {round(numeric_slope, 6)}")

xs = [i / 20 for i in range(-10, 41)]    # -0.5 … 2.0
ys = [composed(x) for x in xs]
tangent = [composed(x0) + chain_slope * (x - x0) for x in xs]

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(xs, ys, color="tab:blue", label="y = (2x)²")
ax.plot(xs, tangent, "--", color="tab:orange", label="tangent at x=1 (slope 8)")
ax.scatter([x0], [composed(x0)], color="tab:red", zorder=5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("The chain rule gives the composition's slope")
ax.legend()
plt.show()
```

```python
def chain(outer_slope, inner_slope):
    return outer_slope * inner_slope

run_tests([
    ("2 outer, 3 inner", chain(2, 3), 6),
    ("a zero outer slope kills it", chain(5, 0), 0),
    ("a negative outer slope flips sign", chain(-1, 4), -4),
])
```
