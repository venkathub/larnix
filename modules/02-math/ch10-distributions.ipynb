```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)        # fixed seed → the same draws every run
samples = rng.normal(size=1000)       # 1000 draws from the standard normal

fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(samples, bins=30, color="tab:blue", edgecolor="white")
ax.set_xlabel("value")
ax.set_ylabel("count (samples per bin)")
ax.set_title("1000 normal samples → a bell")
plt.show()

print("mean:", round(float(samples.mean()), 3))
print("most extreme draws:", round(float(samples.min()), 2),
      "to", round(float(samples.max()), 2))
```

```python
def relative_frequency(count, total):
    return count / total

run_tests([
    ("quarter of the samples", relative_frequency(250, 1000), 0.25),
    ("half", relative_frequency(1, 2), 0.5),
    ("empty bin", relative_frequency(0, 5), 0.0),
])
```

```python
def in_range(values, low, high):
    count = 0
    for v in values:
        if low <= v <= high:
            count += 1
    return count

run_tests([
    ("middle three", in_range([1, 2, 3, 4, 5], 2, 4), 3),
    ("none in range", in_range([1, 2, 3], 5, 9), 0),
    ("all on the edge", in_range([0, 0, 0], 0, 0), 3),
])
```
