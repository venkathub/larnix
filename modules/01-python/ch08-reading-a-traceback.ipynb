```python
def run_safely(label, func):
    try:
        func()
    except Exception as e:
        print(label, "raised", type(e).__name__, "->", e)

run_safely("int('8h')", lambda: int("8h"))
run_safely("missing key", lambda: {"steps": 7500}["stpes"])
run_safely("divide by zero", lambda: 1 / 0)
```

```python
import traceback

try:
    rows = []
    total = 21000
    total / len(rows)          # len(rows) is 0 -> ZeroDivisionError
except ZeroDivisionError:
    report = traceback.format_exc()

print(report.strip())
print("\nLast line says:", report.strip().splitlines()[-1])
```

```python
def failure_type(func):
    try:
        func()
    except Exception as e:
        return type(e).__name__
    return "no error"

run_tests([
    ("bad int is ValueError", failure_type(lambda: int("x")), "ValueError"),
    ("divide by zero", failure_type(lambda: 1 / 0), "ZeroDivisionError"),
    ("clean call", failure_type(lambda: 5), "no error"),
])
```
