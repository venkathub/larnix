```python
def distinct_moods(days):
    moods = set()
    for d in days:
        moods.add(d["mood"])
    return moods

run_tests([
    ("collects unique moods", distinct_moods([{"mood": "good"}, {"mood": "ok"}, {"mood": "good"}]), {"good", "ok"}),
    ("a single mood", distinct_moods([{"mood": "bad"}]), {"bad"}),
    ("no days, empty set", distinct_moods([]), set()),
])
```
