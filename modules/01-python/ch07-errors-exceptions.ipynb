```python
messy_steps = ["7500", "7,500", "8h", "", "8200"]

cleaned = []
errors = 0
for value in messy_steps:
    try:
        cleaned.append(int(value))
    except ValueError:
        cleaned.append(None)   # couldn't parse this one — mark it
        errors += 1

print("Cleaned:", cleaned)
print("Could not parse", errors, "value(s)")
```

```python
def parse_steps(value):
    try:
        return int(value)
    except ValueError as e:
        print("  skipping", repr(value), "->", e)
        return None

for v in ["8200", "8h"]:
    print(v, "becomes", parse_steps(v))
```

```python
def safe_int(text):
    try:
        return int(text)
    except ValueError:
        return None

run_tests([
    ("parses a number", safe_int("7500"), 7500),
    ("bad input becomes None", safe_int("8h"), None),
    ("empty becomes None", safe_int(""), None),
])
```

```python
def clean_steps(text):
    try:
        return int(text.replace(",", ""))
    except ValueError:
        return 0

run_tests([
    ("plain number", clean_steps("8200"), 8200),
    ("comma is removed", clean_steps("7,500"), 7500),
    ("unparsable becomes 0", clean_steps("8h"), 0),
    ("blank becomes 0", clean_steps(""), 0),
])
```
