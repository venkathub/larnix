```python
GOAL = 8000   # daily step target

def is_active(steps):
    """A day is active when it reaches the step goal."""
    return steps >= GOAL

day_steps = [8200, 5100, 9300, 6000, 8000]
active_days = sum(1 for steps in day_steps if is_active(steps))
percent_active = round(active_days / len(day_steps) * 100)

print((active_days, percent_active))
```

```python
def average_steps(day_steps):
    if not day_steps:
        return 0
    return sum(day_steps) / len(day_steps)

run_tests([
    ("mean of two days", average_steps([8000, 9000]), 8500),
    ("empty list is 0", average_steps([]), 0),
])
```
