```python
def label_day(steps):
    """Label a day by its step count."""
    if steps >= 8000:
        return "active"
    elif steps >= 6000:
        return "ok"
    return "low"

def summarise(days):
    """Print each day's label and return how many were active."""
    active = 0
    for date, steps in days:
        lbl = label_day(steps)        # reuse the function above
        if lbl == "active":
            active += 1
        print(date, "->", lbl)
    return active

days = [("2026-01-01", 7500), ("2026-01-02", 8200), ("2026-01-03", 9100)]
print("Active days:", summarise(days))
```

```python
def met_goal(steps, goal=8000):
    """Did the day reach the goal? Goal defaults to 8000."""
    return steps >= goal

print("8200 vs default goal:", met_goal(8200))
print("5000 vs a 4000 goal:", met_goal(5000, 4000))
```
