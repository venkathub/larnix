```python
import csv

subset = [{"date": r["date"], "steps": r["steps"]} for r in rows[:3]]
with open("subset.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["date", "steps"])
    writer.writeheader()
    writer.writerows(subset)

with open("subset.csv") as f:
    print(f.read())
```

```python
def steps_as_int(row):
    return int(row["steps"])

run_tests([
    ("converts steps", steps_as_int({"date": "x", "steps": "7500"}), 7500),
    ("another row", steps_as_int({"steps": "8200"}), 8200),
])
```

```python
def total_steps(rows):
    total = 0
    for row in rows:
        total += int(row["steps"])
    return total

run_tests([
    ("sums the steps", total_steps([{"steps": "7500"}, {"steps": "8200"}]), 15700),
    ("a single row", total_steps([{"steps": "100"}]), 100),
    ("no rows is zero", total_steps([]), 0),
])
```
